The infectious rate, $\beta$, controls the rate of spread which represents the probability of transmitting disease between a susceptible and an infectious individual. 

The incubation rate, $\sigma$, is the rate of latent individuals becoming infectious (average duration of incubation is $1/\sigma$). 

Recovery rate, $\gamma$ = 1/D, is determined by the average duration, D, of infection. 

For the SEIRS model, $\xi$ is the rate which recovered individuals return to the susceptible statue due to loss of immunity.

$\mu$ and $\nu$ represent the birth and death rates, respectively, and are assumed to be equal to maintain a constant population

```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

scipy.integrate.solve_ivp(fun, t_span, y0, method='RK45', t_eval=None, dense_output=False, events=None, vectorized=False, args=None, **options)

```python
S0=8.6e6 # NYC
beta=0.9 # new exposure = infected*beta*susceptible/total population
sigma=1/7 # a mean of 7 days that a infected person start symptoms
gamma= 3/20  # this beta means in the course of 20 days, given R_0 =3, the average number of people one can affect every day
epsilon=0 
mu=0
nu=0
mortal_rate =0
 
E0=10
I0=1
R0=0
N0=S0+E0+I0+R0
N=N0

y0=[S0,E0,I0,R0]

# S=8.6e2 # NY
# E=1
# I=0
# R=0
# N=S+E+I+R

max_steps=10000

# def fun(t,y):
#     dSdt = mu*N - beta*S*I/N + epsilon*R - nu*S
#     dEdt = beta*S*I/N -sigma*E -mu*E
#     dIdt = sigma*E -gamma*I -nu*I
#     dRdt = gamma*I -epsilon*R -nu*R
#     return [dSdt,dEdt,dIdt,dRdt]


def fun(t,y):
    dSdt = mu*N - beta*y[0]*y[2]/N + epsilon*y[3] - nu*y[0]
    dEdt = beta*y[0]*y[2]/N -sigma*y[1] -nu*y[1]
    dIdt = sigma*y[1] -gamma*y[2] -nu*y[2]
    dRdt = gamma*y[2] -epsilon*y[3] -nu*y[3]
    return [dSdt,dEdt,dIdt,dRdt]

# def fun(t,y):
#     dSdt = mu*N - beta*min(y[1]*50,y[0])                          -nu*y[0]           +epsilon*y[3]
#     dEdt =        beta*min(y[1]*50,y[0]) - sigma*y[1]              -nu*y[1]
#     dIdt =                                 sigma*y[1] -gamma*y[2]    -nu*y[2]
#     dRdt =                                          gamma*y[2]     -nu*y[3] -mortal_rate*y[2] -epsilon*y[3]
#     return [dSdt,dEdt,dIdt,dRdt]

```

```python
ans=solve_ivp(fun,[0,300],y0)
```

```python
# ans
```

```python
deaths = N0-np.array(ans.y[0])-np.array(ans.y[1])-np.array(ans.y[2])-np.array(ans.y[3])
culmulated_infection = np.array(ans.y[1]) +np.array(ans.y[2]) +np.array(ans.y[3])
times_1000 = np.where(ans.y[2]>1000)[0][0]
print('day',int(ans.t[times_1000-4]), '\tinfected', int(ans.y[2][times_1000-4]),'\tdeaths', int(deaths[times_1000-4])  )
print('day',int(ans.t[times_1000-2]), '\tinfected', int(ans.y[2][times_1000-2]),'\tdeaths', int(deaths[times_1000-2])  )
print('day',int(ans.t[times_1000]), '\tinfected', int(ans.y[2][times_1000]),'\tdeaths', int(deaths[times_1000])  )
print('day',int(ans.t[times_1000+3]), '\tinfected', int(ans.y[2][times_1000+3]),'\tdeaths', int(deaths[times_1000+3])  )
print( 'total deaths', int (deaths[-1]))

plt.figure(figsize=(10,3))
plt.plot(ans.t, ans.y[0],c='cyan')
plt.plot(ans.t, ans.y[1],c='gold')
plt.plot(ans.t, ans.y[2],c='r')
plt.plot(ans.t, ans.y[3],c='chartreuse')
plt.plot(ans.t, deaths,c='k',linestyle='--')
plt.plot(ans.t, culmulated_infection,linestyle='dashed')

plt.xlabel('t')
# plt.yscale('log')
# plt.ylim([1,S0])
plt.legend(['Susceptible', 'Exposed','Infective','Resistant','Death','total_infected'], shadow=True)
plt.title('SEIR model')
plt.show()
```

# consider patient dies

```python
S0=8.6e6 # NYC
beta= 0.015 ##changed:  new exposure = infected/total population*beta* min(susceptible,150*infected)
          
sigma=1/7 # a mean of 7 days that a infected person start symptoms
gamma= 3/20  # this beta means in the course of 20 days, given R_0 =3, the average number of people one can affect every day
epsilon=0 
mu=0
nu=0
mortal_rate = 0.03/20 # 3% infected people die after 20 days of symptom

E0=10
I0=1
R0=0
N0=S0+E0+I0+R0
N=N0

y0=[S0,E0,I0,R0]

# S=8.6e2 # NY
# E=1
# I=0
# R=0
# N=S+E+I+R
 
max_steps=10000


def fun(t,y):
    dSdt = mu*N - beta*min(y[1]*50,y[0])                          -nu*y[0]           +epsilon*y[3]
    dEdt =        beta*min(y[1]*50,y[0]) - sigma*y[1]              -nu*y[1]
    dIdt =                                 sigma*y[1] -gamma*y[2]    -nu*y[2]
    dRdt =                                          gamma*y[2]     -nu*y[3] -mortal_rate*y[2] -epsilon*y[3]
    return [dSdt,dEdt,dIdt,dRdt]
```

```python
ans=solve_ivp(fun,[0,300],y0)
```

```python
deaths = N0-np.array(ans.y[0])-np.array(ans.y[1])-np.array(ans.y[2])-np.array(ans.y[3])
culmulated_infection = np.array(ans.y[1]) +np.array(ans.y[2]) +np.array(ans.y[3])
times_1000 = np.where(ans.y[2]>1000)[0][0]
print('day',int(ans.t[times_1000-4]), '\tinfected', int(ans.y[2][times_1000-4]),'\tdeaths', int(deaths[times_1000-4])  )
print('day',int(ans.t[times_1000-2]), '\tinfected', int(ans.y[2][times_1000-2]),'\tdeaths', int(deaths[times_1000-2])  )
print('day',int(ans.t[times_1000]), '\tinfected', int(ans.y[2][times_1000]),'\tdeaths', int(deaths[times_1000])  )
print('day',int(ans.t[times_1000+3]), '\tinfected', int(ans.y[2][times_1000+3]),'\tdeaths', int(deaths[times_1000+3])  )
print( 'total deaths', int (deaths[-1]))

plt.figure(figsize=(10,3))
plt.plot(ans.t, ans.y[0],c='cyan')
plt.plot(ans.t, ans.y[1],c='gold')
plt.plot(ans.t, ans.y[2],c='r')
plt.plot(ans.t, ans.y[3],c='chartreuse')
plt.plot(ans.t, deaths,c='k',linestyle='--')
plt.plot(ans.t, culmulated_infection,linestyle='dashed')

plt.xlabel('t')
# plt.yscale('log')
# plt.ylim([1,S0])
plt.legend(['Susceptible', 'Exposed','Infective','Resistant','Death','total_infected'], shadow=True)
plt.title('SEIR model')
plt.show()
```

# consider virus mutation

```python
S0=8.6e6 # NYC
beta= 0.015  ##changed:  new exposure = infected/total population*beta* min(susceptible,150*infected)
          
sigma=1/7 # a mean of 7 days that a infected person start symptoms
gamma= 3/20  # this beta means in the course of 20 days, given R_0 =3, the average number of people one can affect every day
xi= 1/200/1e3  # observe 1 mutation every 200 days, among a population of 10000 infected people
# xi=0
mu=0
nu=0
mortal_rate = 0.03/20 # 3% infected people die after 20 days of symptom

E0=10
I0=1
R0=0
N0=S0+E0+I0+R0
N=N0

y0=[S0,E0,I0,R0]

# S=8.6e2 # NY
# E=1
# I=0
# R=0
# N=S+E+I+R
 
max_steps=10000


def fun(t,y):
    dSdt = mu*N - beta*min(y[1]*50,y[0])                          -nu*y[0]                   +xi*y[3]
    dEdt =        beta*min(y[1]*50,y[0]) - sigma*y[1]              -nu*y[1]
    dIdt =                                 sigma*y[1] -gamma*y[2]    -nu*y[2]
    dRdt =                                          gamma*y[2]     -nu*y[3] -mortal_rate*y[2] -xi*y[3]
    return [dSdt,dEdt,dIdt,dRdt]
```

```python
ans=solve_ivp(fun,[0,300],y0)
```

```python
deaths = N0-np.array(ans.y[0])-np.array(ans.y[1])-np.array(ans.y[2])-np.array(ans.y[3])
culmulated_infection = np.array(ans.y[1]) +np.array(ans.y[2]) +np.array(ans.y[3])
times_1000 = np.where(ans.y[2]>1000)[0][0]
print('day',int(ans.t[times_1000-4]), '\tinfected', int(ans.y[2][times_1000-4]),'\tdeaths', int(deaths[times_1000-4])  )
print('day',int(ans.t[times_1000-2]), '\tinfected', int(ans.y[2][times_1000-2]),'\tdeaths', int(deaths[times_1000-2])  )
print('day',int(ans.t[times_1000]), '\tinfected', int(ans.y[2][times_1000]),'\tdeaths', int(deaths[times_1000])  )
print('day',int(ans.t[times_1000+3]), '\tinfected', int(ans.y[2][times_1000+3]),'\tdeaths', int(deaths[times_1000+3])  )
print( 'total deaths', int (deaths[-1]))

plt.figure(figsize=(10,3))
plt.plot(ans.t, ans.y[0],c='cyan')
plt.plot(ans.t, ans.y[1],c='gold')
plt.plot(ans.t, ans.y[2],c='r')
plt.plot(ans.t, ans.y[3],c='chartreuse')
plt.plot(ans.t, deaths,c='k',linestyle='--')
plt.plot(ans.t, culmulated_infection,linestyle='dashed')

plt.xlabel('t')
plt.legend(['Susceptible', 'Exposed','Infective','Resistant','Death','total_infected'], shadow=True)
plt.title('SEIR model')
plt.show()
```
